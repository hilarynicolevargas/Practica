```
import pandas as pd
import numpy as np
```

```
basket = pd.read_csv('basketball.csv')
```

```
basket.head()
```

```
y=basket.success_field_goals
x=basket.height
```

```
m, b = np.polyfit(x,y,1)
```

```
y_mean = y.mean()
x_mean = x.mean()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(x,y, 'o', markersize = 14, label = "Tiros de campo exitoso para un jugador")
plt.plot(x_mean,y_mean, '*', markersize = 30, color = "r")
plt.plot(x, m*x + b, '-', label = 'Regresion lineal simple', linewidth = 4)
plt.ylabel('Porcentaje de tiros de campo exitoso', fontsize = 20)
plt.xlabel('Altura', fontsize = 30)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.annotate('Centroide', xy=(x_mean-0.1, y_mean-5), xytext=(x_mean-3, y_mean-20), arrowprops=dict(facecolor='black', shrink=0.05), fontsize = 30)
plt.legend(loc = 'upper right', fontsize = 20)
plt.show()
```

```
x_input=basket.drop(['success_field_goals'], axis=1)
```

```
x_input.head()
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_input,y, test_size=0.20)
```

```
x_train.info()
```

```
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```
reg.fit(x_train,y_train)
```

```
pred = reg.predict(x_test)
```

```
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```
